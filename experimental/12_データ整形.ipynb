```python
import pandas as pd
import numpy as np
from sklearn.neighbors import BallTree
```

```python
datarate = pd.read_csv('../results/22_data_rate.csv')
datarate.describe().to_csv('../results/22_data_rate_describe.csv', encoding='utf-8-sig')
```

```python
hinanzyo = pd.read_csv('../data/mergeFromCity_1.csv')
hinanzyo
```

```python
byouin = pd.read_csv('../results/22_医療機関データ_latlon追加.csv')
byouin
```

```python
# クライアントとシェルターの抽出
clients = byouin[['P04_002','lat','lng']].copy()
shelters = datarate[['MESH_ID','lat','lon']].copy()

# ---- 緯度経度をラジアンに変換 ----
clients_rad = np.radians(clients[['lat','lng']].values)
shelters_rad = np.radians(shelters[['lat','lon']].values)

# ---- BallTree を作成（haversine距離）----
tree = BallTree(clients_rad, metric='haversine')

# ---- 1番近いシェルターを探索 ----
dist, ind = tree.query(shelters_rad, k=1)

# ---- 結果を結合 ----
# haversineはラジアン単位なので地球半径をかけて距離[km]に変換
earth_radius = 6371.0
shelters['近隣病院'] = clients.iloc[ind.flatten()]['P04_002'].values
shelters['病院'] = dist.flatten() * earth_radius#km
shelters
```

```python
# datarate と結合（MESH_IDをキーに）
datarate = datarate.merge(
    shelters[["MESH_ID",'近隣病院','病院']],                # Seriesなので自動的に DataFrame 扱いされる
    left_on="MESH_ID",    # datarate 側のキー
    right_on="MESH_ID",     # A_exp 側のキーは index
    #how="left"            # 一致しない場合も残したいなら left
)
#datarate
```

```python
# 都道府県名、市町村名のどちらかに「静岡県」を含む行を抽出
mask = hinanzyo['都道府県名及び市町村名'].str.contains("静岡県", na=False)
hinanzyo = hinanzyo[mask].copy()
```

```python
# ---- 緯度経度をラジアンに変換 ----
clients = hinanzyo[['共通ID','緯度','経度']].copy()
clients_rad = np.radians(clients[['緯度','経度']].values)
shelters_rad = np.radians(shelters[['lat','lon']].values)

# ---- BallTree を作成（haversine距離）----
tree = BallTree(clients_rad, metric='haversine')

# ---- 1番近いシェルターを探索 ----
dist, ind = tree.query(shelters_rad, k=1)

# ---- 結果を結合 ----
# haversineはラジアン単位なので地球半径をかけて距離[km]に変換
earth_radius = 6371.0
shelters['近隣避難所'] = clients.iloc[ind.flatten()]['共通ID'].values
shelters['避難所'] = dist.flatten() * earth_radius#km

# datarate と結合（MESH_IDをキーに）
datarate = datarate.merge(
    shelters[["MESH_ID",'近隣避難所','避難所']],                # Seriesなので自動的に DataFrame 扱いされる
    left_on="MESH_ID",    # datarate 側のキー
    right_on="MESH_ID",     # A_exp 側のキーは index
    #how="left"            # 一致しない場合も残したいなら left
)
#datarate
```

```python
#カラム名
datarate.columns.to_list()
```

```python
#津波
# ラベル → 下端値 の辞書
tsunami_label = {
    '津波浸水想定データ_0.01m以上 〜 0.3m未満': 0.3,
    '津波浸水想定データ_0.3m以上 〜 1m未満': 1,
    '津波浸水想定データ_1m以上 〜 2m未満': 2,
    '津波浸水想定データ_2m以上 〜 3m未満': 3,
    '津波浸水想定データ_3m以上 〜 5m未満': 5,
    '津波浸水想定データ_5m以上 〜 10m未満': 10,
    '津波浸水想定データ_10m以上 〜 20m未満': 20,
    '津波浸水想定データ_20m以上': 30
}

# 各列の「下端×面積」を足し合わせて新しい列に
datarate["津波"] = sum(datarate[col] * tsunami_label[col] for col in tsunami_label if col in datarate.columns)
datarate
```

```python
#洪水
kouzui_label = {
    1:0.5,#0m以上0.5m未満
    2:3,#0m以上0.5m未満
    3:5,#3.0m以上5.0m未満
    4:10,#5.0m以上10.0m未満
    5:20,#10.0m以上20.0m未満
    6:30#20.0m以上
}

# 各列の「下端×面積」を足し合わせて新しい列に
datarate["洪水"] = sum(datarate[f'洪水浸水想定区域データ（河川単位）_20_想定最大規模_{col}'] * kouzui_label[col] for col in kouzui_label if f'洪水浸水想定区域データ（河川単位）_20_想定最大規模_{col}' in datarate.columns)
datarate
```

```python
#土砂災害
dosya_label = {
    '土砂災害警戒区域データ_区域区分_1':1,
    '土砂災害警戒区域データ_区域区分_2':5
    }

# 各列の「下端×面積」を足し合わせて新しい列に
datarate["土砂"] = sum(datarate[col] * dosya_label[col] for col in dosya_label if col in datarate.columns)
datarate
```

```python
#地震ハザード
datarate['地震'] = datarate['T30_I60_PS_max'].copy()
datarate
```
